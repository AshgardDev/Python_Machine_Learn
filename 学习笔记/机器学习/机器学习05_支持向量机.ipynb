```python
from sklearn.svm import SVR
import numpy as np

np.random.seed(1234)
X = np.random.randint(1, 10, size=(500, 2))
## y = 2*x_0 + 3*x_1 + 4
y = np.dot(X, np.array([2, 3])) + 4
# model = SVR(kernel='linear')
# model = SVR()
# svm包括三个参数:核函数及其系数、惩罚参数
model = SVR(kernel='rbf', gamma=1, C=50, epsilon=0.01)
model.fit(X, y)
# print('svr模型系数是:', model.coef_)
# print('svr模型偏置是:', model.intercept_)
pred = model.predict(np.array([[4, 5]]))
print("svr模型预测值:", pred)
print("svr模型评分值:", model.score(X, y))
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

iris = datasets.load_iris()
param_grid = {'kernel': ('rbf', 'linear'), 'C': [1, 10]}
svc = svm.SVC()
### n_jobs代表2核并行 -1时代表不限制cpu核数
clf = GridSearchCV(svc, param_grid, n_jobs=2)
clf.fit(iris.data, iris.target)
```

```python
## svm算法
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import sklearn.svm as svm

X, y = sk.datasets.load_iris(return_X_y=True)
X = X[:, :2]
model = svm.SVR(kernel='rbf', C=1, gamma=1, epsilon=0.01)
model.fit(X, y)

# print(model.coef_)
# print(model.intercept_)
print(model.support_vectors_)

n = 1000
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, n), np.linspace(y_min, y_max, n))
zz = model.predict(np.c_[xx.ravel(), yy.ravel()])

plt.figure(figsize=(8, 8))

plt.contourf(xx, yy, zz.reshape(xx.shape), alpha=0.3)
colors = ['red', 'green', 'blue']
for i, color in zip([0, 1, 2], colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, edgecolor='k')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```

```python
## svm算法
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import sklearn.svm as svm

X, y = sk.datasets.load_iris(return_X_y=True)
X = X[:, :2]
model = svm.SVR(kernel='linear', C=1, gamma=1, epsilon=0.01)
model.fit(X, y)

# print(model.coef_)
# print(model.intercept_)
# print(model.support_vectors_)

w = model.coef_[0]
b = model.intercept_[0]   #训练后分类面的参数 w[0]*x[0] + w[1]*x[1] + b = 0
s_0 = model.support_vectors_[0]   #负类的支持向量
s_1 = model.support_vectors_[1]   #正类的支持向量

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
x = np.linspace(x_min, x_max, n)    #画分类面
svc_y = (-x*w[0] - b)/w[1]

b_0 = -(w[0]*s_0[0] + w[1]*s_0[1])   #穿过负类支撑向量的直线
f_0 = (-x*w[0] - b_0)/w[1]

b_1 = -(w[0]*s_1[0] + w[1]*s_1[1])    # 穿过正类支撑向量的线
f_1 = (-x*w[0] - b_1)/w[1]

plt.figure(figsize=(8, 8))

plt.plot(x, svc_y, 'r-', label='svr')
plt.plot(x, f_0, 'b-', label='b')
plt.plot(x, f_1, 'g-', label='f')

# plt.contourf(xx, yy, zz.reshape(xx.shape), alpha=0.3)
colors = ['red', 'green', 'blue']
for i, color in zip([0, 1, 2], colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, edgecolor='k')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```
