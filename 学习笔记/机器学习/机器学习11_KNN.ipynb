```python
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
## knn 算法
X, y = sk.datasets.load_iris(return_X_y=True)
X = X[:, :2]
n_neighbors = 8
knn_model = sk.neighbors.KNeighborsClassifier(n_neighbors)
knn_model.fit(X, y)

n = 1000
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, n), np.linspace(y_min, y_max, n))
zz = knn_model.predict(np.c_[xx.ravel(), yy.ravel()])


plt.figure(figsize=(8, 8))
plt.contourf(xx, yy, zz.reshape(xx.shape), alpha=0.3)

colors = ['red', 'green', 'blue']
for i, color in zip([0, 1, 2], colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, edgecolor='k')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```
