```python
## 在matplotlib中怎么使用
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

plt.title(r"y=sin(x)")
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x))
plt.xlabel(r"$x$")
plt.ylabel(r"$\sin(x)$")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r"-$\pi$", r"$\frac{\pi}{2}$", "0", r"$\frac{\pi}{2}$",r"$\pi$"]
           )


plt.show()
```
