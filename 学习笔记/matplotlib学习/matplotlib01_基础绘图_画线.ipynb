```python
x = np.linspace(-2*np.pi, 2* np.pi, 1000)

## 正弦曲线
plt.plot(x, np.sin(x), linewidth=2, linestyle='-', color='b', alpha=0.5)
## 余弦曲线
plt.plot(x, np.cos(x)/2, color='#345678', linestyle=':')

## 控制显示区域--坐标轴范围
# plt.xlim(0, 2*np.pi)
# plt.ylim(0, 2)

## 设置坐标轴刻度
## 第一个数组 坐标轴的刻度值
## 第二个数组 坐标轴刻度值对应的标签文本
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], ['-2π', '-π', '0', 'π', '2π'])
plt.yticks([-1, -0.5, 0, 0.5, 1])


### 更改坐标轴位置

### 获取当前坐标系
# https://matplotlib.org/stable/api/spines_api.html#matplotlib.spines.Spine.set_position
axes = plt.gca()
## 设置坐标轴不可见(脊)
axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
## 移动坐标轴到数字0的位置
# axes.spines['bottom'].set_position(('data', 0))
# axes.spines['left'].set_position('center')
axes.spines['bottom'].set_position(('axes', 0.5))
axes.spines['left'].set_position(('data', 0))
axes.spines['left'].set_color('#987654')

## xaxis坐标轴本身是不能定位自己在哪里的,只能通过他的坐标系axes来定位
## xaxis可以设置刻度、标签、颜色等
axes.xaxis.set_label_text('x轴', color='r')
axes.yaxis.set_label_text('y轴')
axes.xaxis.set_label_position('bottom')

plt.show()

## plt操作是直接作用当前的子图,新建子图后,默认激活,可以通过plt.sca(ax1)切换子图
# Matplotlib 提供了两种使用方式：一种是面向对象的方式，另一种是状态机（stateful）模式。
#
# 状态机模式：
# 当你调用 plt.subplot() 时，Matplotlib 会将创建的子图设为当前的 Axes，然后你使用 plt.plot() 等函数时，会自动作用于当前 Axes。这种方式比较简单，适合交互式绘图或简单的绘图任务。
#
# 面向对象方式：
# 返回的 ax 对象可以让你更明确地控制每个子图的属性，这种方式更灵活且更推荐用于复杂图形绘制。通过 ax.plot() 可以直接指定绘图对象，而不依赖于当前状态。
#
# 因此，即使你使用 plt.subplot() 创建了子图，也可以直接用 plt.plot()，因为 Matplotlib 会将 plt 的函数默认应用到当前活动的 Axes 上。两种方式都可以达到绘图效果，只不过面向对象的方式更清晰、易于维护。
```

```python
## 设置备注点
x = np.linspace(-2*np.pi, 2* np.pi, 1000)

plt.figure()
plt.grid()
## 正弦曲线
plt.plot(x, np.sin(x), linewidth=2, linestyle='-', color='b', alpha=0.5, label='sin(x)')
plt.scatter([np.pi/2],  [np.sin(np.pi/2)], marker='o', color='y',)
#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html#matplotlib.axes.Axes.annotate
plt.annotate('test',
             xy=(0, 0),
             xytext=(0.5, -0.5),
             arrowprops=dict(facecolor='red', arrowstyle='-|>', connectionstyle="arc3,rad=0.2")
             )

plt.annotate("max value0",
             xy=(np.pi / 2, np.sin(np.pi / 2)),
             xytext=(np.pi/ 2, np.sin(np.pi / 2) + 0.25), ## 文本位置,相对于整个坐标系
             #ha='center', ## 控制文本水平居中,相对于xy点
             #va='center', ## 控制文本垂直居中,相对于xy点
             arrowprops=dict(arrowstyle='->')
             )

plt.annotate("max value1",
             xy=(np.pi / 2, np.sin(np.pi / 2)),
             xytext=(-80, 20), ## xy点参照系
             textcoords='offset points', ## 修改文本的坐标系,默认是data,可以设置成相对于xy点 长度计算单位:分辨率
             #ha='center', ## 控制文本水平居中,相对于xy点
             #va='center', ## 控制文本垂直居中,相对于xy点
             arrowprops=dict(arrowstyle='->')
             )

## 刻度标签
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], ['-2π', '-π', '0', 'π', '2π'])
plt.yticks([-1, -0.5, 0, 0.5, 1])

plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2, 2)

## 设置图例,必须要线设置绘图的图形的label, 如上面两条线
plt.legend(loc='upper right') ## loc可以设置0-10,具体查文档
plt.show()

```

```python
### 图像对象
fig = plt.figure(
    'A',
    figsize=(4, 3),
    dpi=80,
    facecolor='w',
    edgecolor='k',
)
plt.plot(np.random.rand(5), np.random.rand(5), color='r')

fig2 = plt.figure('B', figsize=(4, 3))
### 这里plt操作的将会是第二个figure。 fig B
plt.scatter(np.random.rand(10), np.random.rand(10), color='b')

## 执行这句代码,又切换回fig A
plt.figure('A')
plt.scatter(np.random.rand(10), np.random.rand(10), color='g')
plt.show()
```
