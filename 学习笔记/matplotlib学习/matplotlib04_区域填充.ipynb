```python
### 填充
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['STHeiti']
plt.rcParams['axes.unicode_minus'] = False
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.fill(x, y, alpha=0.3, color='orange') ## 默认填充是与x轴的封闭区间
plt.show()
```

```python
x1 = np.linspace(0, 10 * np.pi, 100)
y1 = np.sin(x1)
y2 = np.cos(x1)

plt.plot(x1, y1, label='sin(x)')
plt.plot(x1, y2, label='cos(x)')

## 填充2个y轴的封闭区间
## y1 >= y2是可选条件
plt.fill_between(x1, y1, y2, np.sin(x1) > np.cos(x1), color='y', alpha=0.3)

plt.legend()
plt.show()
```
