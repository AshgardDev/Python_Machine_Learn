```python
##条形图
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 13, 1)
y = np.random.normal(100, 20, size=12)

plt.figure()
plt.bar(
    x,
    height=y,
    width=0.8, ## 默认值是1,占比100%
    bottom=0.0,
    align='center', ## 相对于当前x的刻度值
    color='b',
    alpha=0.3,
    edgecolor='k',
)

ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(1))

plt.show()
```

```python
## 重叠
month = np.arange(1, 13)
apples_sale=np.random.normal(100, 20, size=12)
oranges_sale=np.random.normal(150, 40, size=12)
plt.figure()

plt.bar(month, apples_sale, label='apples sale', align='center', width=0.8)
plt.bar(month, oranges_sale, label='oranges sale', align='center', width=0.8)

plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.tick_params(labelsize=10)
plt.grid(linestyle=':')
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```

```python
## 偏移量设置

## 设置中文编码
plt.rcParams['font.sans-serif'] = ['STHeiti']
## 设置-号
plt.rcParams['axes.unicode_minus'] = False

month = np.arange(1, 13)
apples_sale=np.random.normal(100, 20, size=12)
oranges_sale=np.random.normal(150, 40, size=12)
plt.figure()

## 注意计算偏移量
plt.bar(month -0.2 , apples_sale, label='apples sale', align='center', width=0.4)
plt.bar(month + 0.2, oranges_sale, label='oranges sale', align='center', width=0.4)

# 等价
# plt.bar(x - 0.4, apples_sale, label='apples sale', align='edge', width=0.4)
# plt.bar(x, oranges_sale, label='oranges sale', align='edge', width=0.4)


plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.tick_params(labelsize=10)
plt.grid(linestyle=':')
ax = plt.gca()
ax.tick_params(labelsize=10, rotation=45)
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{_}月"))
plt.legend()
plt.show()
```
