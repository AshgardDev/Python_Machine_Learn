```python
import matplotlib.pyplot as plt
import numpy as np

n = 300

height = np.random.normal(172, 20, n)
weight = np.random.normal(60, 10, n)

plt.figure('Persons Height&Weight', facecolor='lightgray')
plt.title('Persons Height&Weight', fontsize=16)
plt.xlabel('Height', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.grid(linestyle=':', alpha=0.3)

height_avg = np.average(height)
weight_avg = np.average(weight)
c_distance = (height - height_avg)**2 + (weight - weight_avg)**2
s = np.random.normal(5, 20, n)
plt.scatter(height, weight,
            marker='o',
            # color='yellow', ##当颜色不在是一个值时,而是与数据长度一样的数组时,可以给cmap设置颜色映射
            c = c_distance, # 设置值c,映射cmap
            s= s, # 设置随机的大小
            label='Persons Height&Weight',
            cmap='rainbow'
            #cmap='jet'
            )

plt.legend()
plt.show()
```
