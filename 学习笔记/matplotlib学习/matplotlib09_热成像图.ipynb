```python
### 热成像图 (用图像的方式显示矩阵及矩阵中值的大小)
## 把矩阵z图形化,使用cmap表示矩阵中每个元素值的大小
import numpy as np
import matplotlib.pyplot as plt

n = 100
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
## 说一下meshgrid的作用,其实就是将x,y的坐标生成网格矩阵
## 也就是: x会沿着y轴方向复制len(y)个, y会沿着x轴方向复制len(x)个
## 所以x,y的每个坐标点组合起来就是整个网格的坐标(x,y)
m, n = np.meshgrid(x, y) ## 复制了就会变成2维数组
z = (1 -m/2 + m**5 + n**3) * np.exp(-m**2 - n**2)

## 其实本质上就是二位数组的值,与颜色的映射
# z = np.arange(1, 10).reshape(3, 3)
# print(z)

plt.figure()
plt.imshow(
    z,
    interpolation='nearest', ## 控制图像平滑过渡的方式,插值法--采用最近的
    cmap='jet',
    origin='lower',## 数据原点位置, upper高,lower低
    alpha=0.9,
)

## 使用颜色条显示热度值
plt.colorbar()

plt.show()
```

```python
## 设置中文编码
plt.rcParams['font.sans-serif'] = ['STHeiti']
## 设置-号
plt.rcParams['axes.unicode_minus'] = False

# 生成数据
data = np.sin(np.linspace(0, 10, 100)).reshape(10, 10)

# 绘制图像
plt.imshow(data,
           cmap='coolwarm',        # 颜色映射
           interpolation='bicubic',# 平滑插值
           extent=(0, 10, 0, 10),  # 坐标范围
           vmin=-1, vmax=1,        # 数据范围
           origin='lower',         # 原点在下
           alpha=0.8)              # 透明度
plt.colorbar(label='值')           # 颜色条
plt.title('二维正弦热图')
plt.xlabel('X 轴')
plt.ylabel('Y 轴')
plt.show()
```
