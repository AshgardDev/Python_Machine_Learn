```python
## 饼图
import matplotlib.pyplot as plt
import numpy as np
values = np.array([35, 70, 45, 25, 15])
labels = ['java', 'python', 'go', 'c', 'c#']
spaces = [0.0, 0.05, 0.0, 0.0, 0.0]
plt.figure()
## 等轴比例设置
## Tools -> Python Plots -> 把复选框去掉,就可以独立窗口显示
ax = plt.gca()
ax.set_box_aspect(1)
plt.pie(
    values,
    labels=labels,
    explode=spaces,
    colors=['orange', 'green', 'blue', 'cyan', 'purple'],
    shadow=True,
    # autopct='%.0f%%',
    # autopct='%1.1f',
    autopct='%1.2f%%',
    # autopct=lambda pct: f'{pct:.1f}%',
    startangle=-100,
    radius=0.8,
    pctdistance=0.55,
    labeldistance=1.2
)

plt.show()
```
